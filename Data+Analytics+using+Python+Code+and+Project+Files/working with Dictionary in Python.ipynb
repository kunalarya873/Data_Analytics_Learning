# Dictionary

Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

Python dictionary is an unordered collection of items. While other compound data types have only value as an element, a dictionary has a key: value pair.

# Dict Creation

```python
thisdict = {
  "brand": "Hyundai",
  "model": "Grand i10",
  "year": 2018
}
print(thisdict)
```

**Dictionary Items**

Dictionary items are ordered, changeable, and does not allow duplicates.

Dictionary items are presented in key:value pairs, and can be referred to by using the key name.

```python
thisdict = {
  "brand": "Hyundai",
  "model": "Grand i10",
  "year": 2018
}
print(thisdict["model"])
```

**Ordered or Unordered?**

As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

When we say that dictionaries are ordered, it means that the items have a defined order, and that order will not change.

Unordered means that the items does not have a defined order, you cannot refer to an item by using an index.

**Changeable**

Dictionaries are changeable, meaning that we can change, add or remove items after the dictionary has been created.

Duplicates Not Allowed
Dictionaries cannot have two items with the same key:

```python
#Duplicate values will overwrite existing values:

thisdict = {
  "brand": "Hyundai",
  "model": "Grand i10",
  "year": 2018,
  "year1": 2020
}
print(thisdict)
```

**Dictionary Length**

To determine how many items a dictionary has, use the len() function:

```python
print(len(thisdict))
```

**Dictionary Items - Data Types**

The values in dictionary items can be of any data type:

```python
#String, int, boolean, and list data types:

thisdict = {
   "brand": "Hyundai",
   "model": "Grand i10",
   "year": 2018,
   "year": 2020,
   "colors": ["Fiery Red", "Blue", "Orange"],
   
}
print(thisdict)
```

**The dict() Constructor**

It is also possible to use the dict() constructor to make a dictionary.

Example
Using the dict() method to make a dictionary:

```python
thisdict = dict(name = "Jawahar", age = 42, country = "India", salary= 40000)
print(thisdict)
```

```python
#empty dictionary
thisdict = {}

#dictionary with integer keys
thisdict = {1: 'WWE', 2: 'IPL'}
print(thisdict)

#dictionary with mixed keys
thisdict = {'Name': 'Ankur', 1: ['WWE', 'IPL']}
print(thisdict)


#create empty dictionary using dict()
thisdict = dict()

thisdict = dict([(1, 'WWE'), (2, 'IPL')])    #create a dict with list of tuples
print(thisdict)
```

# Dict Access

```python
thisdict = {'Name': 'Ankur', 'Age': 27, 'City': 'Deoria'}

#get name
print(thisdict['Name'])
```

```python
#if key is not present it gives KeyError
print(thisdict['State'])
```

```python
#another way of accessing key
print(thisdict.get('City'))
```

```python
#if key is not present it will give None using get method
print(thisdict.get('State'))
```

# Dict Add or Modify  Elements

```python
thisdict =  {'Name': 'Ankur', 'Age': 27, 'City': 'Deoria'}

#update name 
thisdict['Name'] = 'Abhay Shankar'

print(thisdict)
```

```python
#add new key
thisdict['State'] = 'Uttar Pradesh'

print(thisdict)
```

# Dict Delete or Remove Element

```python
#create a dictionary
thisdict = {'Name': 'Abhay Shankar', 'Age': 27, 'City': 'Deoria', 'State': 'Uttar Pradesh'}

#remove a particular item
print(thisdict.pop('Age'))

print(thisdict)
```

```python
thisdict = {'Name': 'Abhay Shankar', 'Age': 27, 'City': 'Deoria', 'State': 'Uttar Pradesh'}

#remove an arbitarty key
thisdict.popitem()

print(thisdict)
```

```python
numsquares = {3: 9, 6: 36, 8: 64, 12: 144}

#delete particular key
del numsquares[8]

print(numsquares)
```

```python
#remove all items
numsquares.clear()

print(numsquares)
```

```python
numsquares = {3: 9, 6: 36, 8: 64, 12: 144}

#delete dictionary itself
del numsquares

print(numsquares) #NameError because dict is deleted
```

# Dictionary Methods

```python
numsquares = {3: 9, 6: 36, 8: 64, 12: 144}

thisdict = numsquares.copy()
print(numsquares)
print(thisdict)
```

```python
#fromkeys[seq[, v]] -> Return a new dictionary with keys from seq and value equal to v (defaults to None).
my_subjects = {}.fromkeys(['History', 'Public Speaking', 'Theater'], 0)
print(my_subjects)
```

```python
numsquares = {3: 9, 6: 36, 8: 64, 12: 144}
print(numsquares.items()) #return a new view of the dictionary items (key, value)
```

```python
numsquares = {3: 9, 6: 36, 8: 64, 12: 144}
print(numsquares.keys()) #return a new view of the dictionary keys
```

```python
numsquares = {3: 9, 6: 36, 8: 64, 12: 144}
print(numsquares.values()) #return a new view of the dictionary values
```

```python
#get list of all available methods and attributes of dictionary
mydict = {}
print(dir(mydict))
```

# Dict Comprehension

```python
#Dict comprehensions are just like list comprehensions but for dictionaries

mydict = {'X': 24, 'Y': 28, 'Z': 32}
for pair in mydict.items():
    print(pair)
```

```python
#Creating a new dictionary with only pairs where the value is larger than 24
mydict = {'X': 24, 'Y': 28, 'Z': 32}
new_dict = {k:v for k, v in mydict.items() if v > 24}
print(new_dict)
```

```python
#We can also perform operations on the key value pairs
mydict = {'X': 24, 'Y': 28, 'Z': 32}
mydict = {k + 'c':v * 2 for k, v in mydict.items() if v > 24}
print(mydict)
```
