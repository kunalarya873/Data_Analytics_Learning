```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
male = sum(df['Sex']=='male')
female = sum(df['Sex']=='female')
print(f"""Males are {male} and Females are {female}""")
```

```python
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.bar(['male', 'female'],[male, female])
```

```python
percent_male = male*100/(male+female)
percent_female = female*100/(male+female)
plt.bar(['male','female'], [percent_male, percent_female])
```

```python
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
male_survived = sum((df['Sex']=='male')& (df['Survived']==1))
female_survived = sum((df['Sex']=='female')& (df['Survived']==1))
labels = ['male', 'female']
not_survived = [male- male_survived, female-female_survived]
survived = [male_survived, female_survived]
plt.bar(labels, survived, color = 'green')
plt.bar(labels, not_survived, bottom= survived, color = 'red')
```

```python
plt.title('Passenger Age')
plt.hist(df['Age'])
```

```python
labels = ['male', 'female']
explode = (0.025, 0.025)# this the slice by which oie chart will offset from each other
sex= [male, female]
plt.pie(sex, explode=explode, labels = labels, autopct ='%.1f%%')# autopct means %age shown in the pieChart, upto one decimal point and add % symbol at the very end of the value
```

```python
x = [1,2,3,4,5] 
y1 = [1, 4, 6, 7, 9] 
y2 = [3, 6, 7, 8, 9] 
y3 = [2, 5, 7, 9, 10]
plt.plot([], [], color='green', label='y1') 
plt.plot([], [], color='red', label='y2') 
plt.plot([], [], color='orange', label='y3')
plt.stackplot(x, y1, y2, y3, colors=['green', 'red', 'orange']) 
plt.legend()
```

```python
x = np.linspace(0, 20, 20)
y = [e*e for e in x] 
plt.subplot(2, 3, 1)
plt.plot(x, y, 'bo-')# Here 'bo-' represents b means blue, o means markus of circle and - means circles are connected with line
```
